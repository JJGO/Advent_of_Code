```python
def simulate(data, days):
    population = dict(Counter(data))
    for i in range(days):
        population = {k-1: v for k, v in population.items()}
        population[8] = population.get(8,0) + population.get(-1,0)
        population[6] = population.get(6,0) + population.pop(-1,0)
    return sum(population.values())
```

```python
def simulate(data, days):
    counts = dict(Counter(data))
    population = [counts.get(i,0) for i in range(9)]
    for i in range(days):
        n = population.pop(0)
        population.append(n)
        population[6] += 1
    return sum(population)
```

```python
def simulate(data, days):
    counts = Counter(data)
    population = deque([counts.get(i,0) for i in range(9)])
    for i in range(days):
        population.roll()
        population[6] += population[-1]
    return sum(population)
```

```python
def fast_simulate(data, days):
    counts = np.zeros(9, dtype=int)
    for i, v in Counter(data).items():
        counts[i] += v
    transition = np.zeros((9,9), dtype=int)
    transition[np.arange(8),np.arange(1,9)] = 1
    transition[6,0] = transition[8,0] = 1
    N = days
    while N > 0:
        N, r = divmod(N, 2)
        if r == 1:
            counts = transition @ counts
        transition = transition @ transition
    return counts.sum()
```
