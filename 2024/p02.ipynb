```python
def is_safe(seq):
    delta = [j-i for i, j in zip(seq, seq[1:])]
    return (all([i>0 for i in delta]) or all([i<0 for i in delta])) and all([ 1 <= abs(i) <= 3 for i in delta])
        
```

```python
[is_safe(l) for l in sample]
```

```python
def solve_a(data):
    return sum(is_safe(line) for line in data)
```

```python
solve_a(sample)
```

```python
solve_a(data)
```

```python
def is_flexsafe(seq):
    if is_safe(seq):
        return True
    for i in range(len(seq)):
        newseq = seq[:]
        newseq.pop(i)
        if is_safe(newseq):
            return True
    return False
```
