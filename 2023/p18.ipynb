```python
def shoelace(points):
    points = np.vstack([points, points[0:1]])
    area = (points[:-1,0] * points[1:,1]).sum() - (points[:-1,1] * points[1:,0]).sum()
    return abs(area)
```

```python
def solve_a(data):
    # shoelace numpy
    V = lambda x, y: np.array([x,y], dtype=np.int64)
    moves = {'R': V(1,0), 'L': V(-1,0), 'U': V(0,1), 'D': V(0,-1)}
    points = np.add.accumulate([V(0,0)]+[n*moves[d] for d, n, _ in data])
    lava = shoelace(points) // 2 + sum((n/2) for _, n, _ in data) + 1
    return int(lava)
```

```python
def solve_a(data):
    # shoelace manual
    x, y = 0, 0
    area, boundary = 0, 0
    moves = {'R': (1,0), 'L': (-1,0), 'U': (0,1), 'D': (0,-1)}
    for move, n, _ in data:
        dx, dy = moves[move]
        x2, y2 = x+n*dx, y+n*dy
        area += 1/2 * (x*y2 - x2*y)
        x, y = x2, y2
        boundary += n
    return int(abs(area) + boundary / 2 + 1)
```

```python
def solve_a(data):
    # green's theorem
    x = 0
    area, boundary = 0, 0
    moves = {'R': (1,0), 'L': (-1,0), 'U': (0,1), 'D': (0,-1)}
    for move, n, _ in data:
        dx, dy = moves[move]
        x += dx * n
        area += x * dy * n
        boundary += n
    return int(abs(area) + boundary / 2 + 1)
```
