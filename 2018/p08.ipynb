```python
from aocd.models import Puzzle

puzzle = Puzzle(year=2018, day=8)

def parses(input):
    return [int(i) for i in input.strip().split()]

data = parses(puzzle.input_data)
```

```python
class Node:
    
    def __init__(self, children, meta):
        self.children = children
        self.meta = meta
    
    def __repr__(self):
        return f'Node({self.children}, {self.meta})'
```

```python
def buildnode(data):
    nchild = data.pop(0)
    nmeta = data.pop(0)
    children = [buildnode(data) for i in range(nchild)]
    meta = [data.pop(0) for i in range(nmeta)]
    return Node(children, meta)
```

```python
sample = parses('2 3 0 3 10 11 12 1 1 0 1 99 2 1 1 2')
```

```python
def tree(data):
    data = data.copy()
    return buildnode(data)
```

```python
def sum_meta(root):
    total = 0
    stack = [root]
    while stack:
        node = stack.pop()
        total += sum(node.meta)
        stack.extend(node.children)
    return total
```

```python
assert sum_meta(tree(sample)) == 138
```

```python
sum_meta(tree(data))
```

```python
puzzle.answer_a = sum_meta(tree(data))
```

```python
from functools import lru_cache
```

```python
@lru_cache(maxsize=None)
def nodeval(node):
    if len(node.children) == 0:
        return sum(node.meta)
    total = 0
    for m in node.meta:
        if m <= len(node.children):
            total += nodeval(node.children[m-1])
    return total
```

```python
assert nodeval(tree(sample)) == 66
```

```python
nodeval(tree(data))
```

```python
puzzle.answer_b = nodeval(tree(data))
```

Iterative

```python
def buildnode(data):
    nchild = data.pop(0)
    nmeta = data.pop(0)
    children = [buildnode(data) for i in range(nchild)]
    meta = [data.pop(0) for i in range(nmeta)]
    return (children, meta)

def tree(data):
    data = data.copy()
    return buildnode(data)

def sum_meta(node):
    children, meta = node
    return sum(meta) + sum(sum_meta(c) for c in children)

def nodeval(node):
    children, meta = node
    if len(children) == 0:
        return sum(meta)
    return sum( nodeval(children[m-1]) for m in meta if m <= len(children))
```

```python
puzzle.answer_a = sum_meta(tree(data))
puzzle.answer_b = nodeval(tree(data))
```
