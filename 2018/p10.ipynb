```python
# Y-axis Variance will decrease until message is spelled, then it will increase again.
# Detect that moment and print message
def sky(coords):
    x, y, vx, vy = np.array(coords).T
    var = np.var(y)
    i = 0
    while True:
        x += vx
        y += vy
        newvar = np.var(y)
        if newvar < var:
            var = newvar
        else:
            x -= vx
            y -= vy
            xmin, xmax = x.min(), x.max() + 1
            ymin, ymax = y.min(), y.max() + 1
            sky = np.zeros((xmax-xmin, ymax-ymin)) 
            sky[x-xmin,y-ymin] = 1
            print('\n'.join("".join(['#' if i else ' ' for i in line]) for line in sky.T ))
            return i
        i += 1
```

```python
def sky(coords):
    # Linear system of equations 
    # x + vx * t = Fx
    # y + vy * t = Fy
    # Solve for t, Fx, Fy
    # Fx, Fy are coords of intersection
    x, y, vx, vy = np.array(coords).T
    N = len(coords)
    B = np.concatenate((x,y)).reshape((2*N,1))
    A = np.zeros((2*N, 3))
    A[:N,0] = -vx
    A[N:,0] = -vy
    A[:N,1] = 1
    A[N:,2] = 1
    sols, *_ = np.linalg.lstsq(A, B)
    t = int(round(sols[0,0]))
    # Print
    x = x + t * vx
    y = y + t * vy
    xmin, xmax = x.min(), x.max() + 1
    ymin, ymax = y.min(), y.max() + 1
    sky = np.zeros((xmax-xmin, ymax-ymin)) 
    sky[x-xmin,y-ymin] = 1
    print('\n'.join("".join(['#' if i else ' ' for i in line]) for line in sky.T ))
    return t
```
